# Imports

```python
import numpy as np
from scipy.linalg import null_space
```

# User Input

```python
m, n = [int(x) for x in input("Enter dimensions of A:").split()]

A = np.zeros((m, n))

# fill up A with values
for i in range(m):
    row = [float(x) for x in input("Enter row " + str(i) + ": ").split()]
    for j in range(n):
        A[i][j] = row[j]
print(A)

# fill up c with values
c = np.array([float(x) for x in input("Enter c: ").split()])
print(c)

# fill up b with values
b = np.array([float(x) for x in input("Enter b: ").split()])
print(b)

# fill up z with values
z = np.array([float(x) for x in input("Enter z: ").split()])
print(z)
```

# Finding a Vertex

```python
while True:
    # declaring required variables
    tight = None
    unTight = None
    tightCount = 0
    unTightCount = 0
    unTightIndices = []

    # dividing A into tight and untight rows
    for i in range(m):
        val = A[i, :] @ z - b[i]

        if abs(val) <= 10 ** -6:
            if tightCount == 0:
                tight = np.copy(A[i, :])
            else:
                tight = np.vstack((tight, A[i, :]))
            tightCount += 1
        else:
            unTightIndices.append(i)
            if unTightCount == 0:
                unTight = np.copy(A[i, :])
            else:
                unTight = np.vstack((unTight, A[i, :]))
            unTightCount += 1
    
    # if the below condition is true, then z = vertex, hence break
    if np.linalg.matrix_rank(tight) == n:
        break
    # update z
    else:
        if(tightCount == 1):
            tight = np.array([tight])
        ns = null_space(tight)
        ns = ns[:, 0]
        
        for i in range(unTightCount):
            unTightRow = unTight[i, :]
            if abs(unTightRow @ ns) <= 10 ** -6:
                continue
            
            dotProd = unTightRow @ ns
            alpha =  (b[unTightIndices[i]] - unTightRow @ z) / dotProd
            z = z + alpha * ns
            break
print(f'Vertex found: ({z[0]},{z[1]})')
```

# Marching through the Vertices

```python
while True:
    # declaring required variables
    tight = None
    unTight = None
    tightCount = 0
    unTightCount = 0
    unTightIndices = []

    # dividing A into tight and untight rows
    for i in range(m):
        val = A[i, :] @ z - b[i]
        if abs(val) <= 10 ** -6:
            if tightCount == 0:
                tight = np.copy(A[i, :])
            else:
                tight = np.vstack((tight, A[i, :]))
            tightCount += 1
        else:
            unTightIndices.append(i)
            if unTightCount == 0:
                unTight = np.copy(A[i, :])
            else:
                unTight = np.vstack((unTight, A[i, :]))
            unTightCount += 1
    
    coeffs = np.linalg.inv(np.transpose(tight)) @ c
    # if the below condition is true, then z = optimum vertex, hence break
    if np.amin(coeffs) >= -1 * (10 ** -6):
        break
    # update z
    else:
        B = np.linalg.inv(tight)
        v = None
        check = False
        
        for i in range(tightCount):
            v = B[:, i]
            if(-v @ c <= 0):
                continue
            
            for j in range(unTightCount):
                try:
                    unTightRow = unTight[j, :]
                except:
                    unTightRow = unTight.copy()
                if abs(unTightRow @ v) <= 10 ** -6:
                    continue
                
                dotProd = unTightRow @ v
                alpha =  -(b[unTightIndices[j]] - unTightRow @ z) / dotProd
                z = z - alpha * v
                check = True
                break
            
            if check:
                break
```

```python
print(f'Final solution: ({z[0]},{z[1]})')
print('Optimal value of objective:', np.dot(c, z))
```
