### Question 1

````
import numpy as np
import matplotlib.pyplot as plt

# taking input from user
points = int(input("Enter the number of points in your signal: ")) 
````

##### a) Plotting a real signal

````
# function to plot a real signal
def plotReal(points):
  signal = []
  print("Enter the input signal:")
  for i in range(points):
    x = int(input())
    signal.append(x)
  signal = np.array(signal)


  #plotting the signal
  plots, output = plt.subplots()
  output.set_title('Real Signal')
  output.set(xlabel='Time [n]', ylabel='Magnitude')
  output.stem(signal, use_line_collection = True) 
  plots.show()

plotReal(points)
````

##### b) Plotting a complex signal

````
#function to plot a complex signal
def plotComplex(points):
  # storing the magnitude and the phase of a complex signal separately
  magnitude = []
  phase = []
  print("Enter the magnitude and the phase of signal:")
  for i in range(points):
    x = input()
    x = x.split()

    magnitude.append(int(x[0]))
    phase.append(int(x[1]))

  magnitude = np.array(magnitude)
  phase = np.array(phase)

  # plotting the magnitue and phase plots separately
  plots, signal = plt.subplots(2)
  signal[0].set_title('Magnitude Plot')
  signal[0].set(xlabel='Time [n]', ylabel='Magnitude')
  signal[0].stem(magnitude, use_line_collection = True) 

  signal[1].set_title('Phase Plot')
  signal[1].set(xlabel='Time [n]', ylabel='Angle')
  signal[1].stem(phase, use_line_collection = True)

  # using tight_layout to automatically space all the plots without any overlaps
  plots.tight_layout()
  plots.show()


plotComplex(points)
````

##### a) Discrete Delta Signal


````
# discrete delta

x = np.arange(-5, 6)
y = np.zeros(x.size)
# setting the values of the point where x = 0 to 1
y[np.where(x == 0)[0][0]] = 1

plt.title('Discrete Delta Signal')
plt.xlabel('Time [n]')
plt.ylabel('Magnitude')
plt.stem(x, y, use_line_collection = True) 
plt.show()
````

##### b) Unit Step Signal


````
# unit step signal

x = np.arange(-5, 6)
y = np.zeros(x.size)
# setting the values of non-negative x's to 1
y[np.where(x == 0)[0][0]:x.size] = 1

plt.title('Unit Step Signal')
plt.xlabel('Time [n]')
plt.ylabel('Magnitude')
plt.stem(x, y, use_line_collection = True) 
plt.show()
````

##### c) Real Exponential Signal


````
# real exponential signal: a^n

x = np.arange(-10, 11)
# parameter input
a = float(input("Exponential parameter: "))
parameter = np.ones(x.size)
parameter = parameter * a
y = np.power(parameter, x)

plt.title('Real Exponential Signal')
plt.xlabel('Time [n]')
plt.ylabel('a^n')
plt.stem(x, y, use_line_collection = True) 
plt.show()
````

##### d) Real Sinusodial Signal


````
# real sinusodial signal: A * sin(omega * n + phi)

x = np.arange(-50, 51)
# parameter input
A = float(input("Amplitude: "))
omega = float(input("Omega: "))
phi = float(input("Phase: "))

# omega * n + phi
angle = np.ones(x.size)       # angle = 1
angle = omega * angle         # angle = omega
angle = np.multiply(angle, x) # angle = omega * x
angle = angle + phi           # angle = omega * x + phi

# calculating sin
y = A * np.sin(angle)

plt.title('Real Sinusodial Signal')
plt.xlabel('Time [n]')
plt.ylabel('A * sin(omega * n + phi)')
plt.stem(x, y, use_line_collection = True) 
plt.show()
````

##### e) Complex Exponential Signal


````
# complex exponential signal: A * e^j*(sigma + n * omega)
# A * e^j*(sigma + n * omega) = A * cos(omega * n + sigma) + j * A * sin(omega * n + sigma)

x = np.arange(-50, 51)
# parameter input
A = float(input("Amplitude: "))
omega = float(input("Omega: "))
sigma = float(input("Phase: "))

# omega * n + sigma
angle = np.ones(x.size)       # angle = 1
angle = omega * angle         # angle = omega
angle = np.multiply(angle, x) # angle = omega * x
angle = angle + sigma           # angle = omega * x + sigma

# calculating real and complex parts
real_part = A * np.sin(angle)
complex_part = A * np.cos(angle)

# plotting the magnitue and phase plots separately
plots, complex_sinusoid = plt.subplots(2, 2)
plots.suptitle('Real Exponential Signal')
complex_sinusoid[0, 0].set_title('Real Part')
complex_sinusoid[0, 0].set(xlabel='Time [n]', ylabel='Cosine')
complex_sinusoid[0, 0].stem(x, real_part, use_line_collection = True) 

complex_sinusoid[0, 1].set_title('Complex Part')
complex_sinusoid[0, 1].set(xlabel='Time [n]', ylabel='Sine')
complex_sinusoid[0, 1].stem(x, complex_part, use_line_collection = True) 

complex_sinusoid[1, 0].set_title('Magnitude')
complex_sinusoid[1, 0].set(xlabel='Time [n]', ylabel='A')
complex_sinusoid[1, 0].stem(x, A * np.ones(x.size), use_line_collection = True) 

complex_sinusoid[1, 1].set_title('Phase')
complex_sinusoid[1, 1].set(xlabel='Time [n]', ylabel='sigma + n * omega')
complex_sinusoid[1, 1].stem(x, angle, use_line_collection = True) 

# using tight_layout to automatically space all the plots without any overlaps
plots.tight_layout()
````

````
import numpy as np
import matplotlib.pyplot as plt
````

````
x = np.array([0, 0, 0, 0.5, 1, 1, 1, 0.5, 0, 0, 0]) # x[n] = u[n/4]
n = x.size
samples = 100

# plotting input signal
plt.title('Input')
plt.xlabel('Time [n]')
plt.ylabel('Magnitude')
plt.stem(np.arange(x.size), x, use_line_collection = True) 
plt.show()

# -pi < w < pi
omega = np.linspace(-np.pi, np.pi, samples)

X = np.zeros(samples, dtype=complex)
for w in range(samples):
  wn = omega[w] * (np.arange(n))
  real = (x * np.cos(wn)).sum()
  imag = (x * np.sin(wn)).sum()

  # X(e^j*omega[w]) = SUM(x * cos(omega[w])) - j *  SUM(x * sin(omega[w])) = A - jB
  X[w] = real - imag * np.array([1j])
````

##### a) Plotting the magnitude spectrum.

````
# Magnitude Spectrum

# magnitude = sqrt(A^2 + B^2)
magnitude = np.sqrt((X.real)*(X.real) + (X.imag)*(X.imag))

plt.title('Magnitude Spectrum')
plt.xlabel('Omega')
plt.ylabel('Magnitude')
plt.stem(omega, magnitude, use_line_collection = True) 
plt.show()
````

##### b) Ploting the phase spectrum.

````
# Phase Spectrum

# phase = arctan(B/A)
# phase = np.arctan(X.imag/X.real)
phase = np.angle(X)

plt.title('Phase Spectrum')
plt.xlabel('Omega')
plt.ylabel('Radians')
plt.stem(omega, phase, use_line_collection = True) 
plt.show()
````

###### a) Verifying whether the DTFT is conjugate symmetric for real signals.

````
# defining input signal
x = np.array([3,2,1,0,1,2,3])
n = np.arange(x.size)
samples = 100

# plotting input signal
plt.title('Input')
plt.xlabel('Time [n]')
plt.ylabel('Magnitude')
plt.stem(np.arange(x.size), x, use_line_collection = True) 
plt.show()
````

````
# DTFT

# -pi < w < pi
omega = np.linspace(-np.pi, np.pi, samples)

X = np.zeros(samples, dtype=complex)
for w in range(samples):
  wn = omega[w] * n
  real = (x * np.cos(wn)).sum()
  imag = (x * np.sin(wn)).sum()

  # X(e^j*omega[w]) = SUM(x * cos(omega[w])) - j *  SUM(x * sin(omega[w])) = A - jB
  X[w] = real - imag * np.array([1j])

X = np.around(X,decimals=5)

# Real part of DTFT

plt.title('Real part of DTFT')
plt.xlabel('Omega')
plt.ylabel('Magnitude')
plt.stem(omega, X.real, use_line_collection = True) 
plt.show()

# Imaginary part of DTFT

plt.title('Imaginary part of DTFT')
plt.xlabel('Omega')
plt.ylabel('Magnitude')
plt.stem(omega, X.imag, use_line_collection = True) 
plt.show()

# Magnitude Spectrum

# magnitude = sqrt(A^2 + B^2)
magnitude = np.sqrt((X.real)*(X.real) + (X.imag)*(X.imag))

plt.title('Magnitude Spectrum')
plt.xlabel('Omega')
plt.ylabel('Magnitude')
plt.stem(omega, magnitude, use_line_collection = True) 
plt.show()

# Phase Spectrum

# phase = arctan(B/A)
phase = np.angle(X)

plt.title('Phase Spectrum')
plt.xlabel('Omega')
plt.ylabel('Radians')
plt.stem(omega, phase, use_line_collection = True) 
plt.show()
````

````
# Calculating X*(-W)

X_conj = np.zeros(samples, dtype=complex)
for w in range(samples):
  wn = -1 * omega[w] * n
  real = (x * np.cos(wn)).sum()
  imag = (x * np.sin(wn)).sum()

  X_conj[w] = real + imag * np.array([1j])

X_conj = np.around(X_conj,decimals=5)

# Real part of DTFT

plt.title('Real part of X*(-W)')
plt.xlabel('Omega')
plt.ylabel('Magnitude')
plt.stem(omega, X_conj.real, use_line_collection = True) 
plt.show()

# Imaginary part of DTFT

plt.title('Imaginary part of X*(-W)')
plt.xlabel('Omega')
plt.ylabel('Magnitude')
plt.stem(omega, X_conj.imag, use_line_collection = True) 
plt.show()

# Magnitude Spectrum

# magnitude = sqrt(A^2 + B^2)
magnitude = np.sqrt((X_conj.real)*(X_conj.real) + (X_conj.imag)*(X_conj.imag))

plt.title('Magnitude Spectrum')
plt.xlabel('Omega')
plt.ylabel('Magnitude')
plt.stem(omega, magnitude, use_line_collection = True) 
plt.show()

# Phase Spectrum

# phase = arctan(B/A)
phase = np.angle(X_conj)

plt.title('Phase Spectrum')
plt.xlabel('Omega')
plt.ylabel('Radians')
plt.stem(omega, phase, use_line_collection = True) 
plt.show()
````

````
if np.array_equal(X,X_conj) == True:
  print("X(W) = X*(-W)")
````

###### We got ```X(W) = X*(-W)```

###### Therefore the DTFT is conjugate symmetric for real signals.

###### b) Verifying whether the DTFT of a real and even signal is real and even.

````
# DTFT

# -pi < w < pi
omega = np.linspace(-np.pi, np.pi, samples)

X = np.zeros(samples, dtype=complex)
for w in range(samples):
  wn = omega[w] * n
  real = (inp * np.cos(wn)).sum()
  imag = (inp * np.sin(wn)).sum()

  # X(e^j*omega[w]) = SUM(x * cos(omega[w])) - j *  SUM(x * sin(omega[w])) = A - jB
  X[w] = real - imag * np.array([1j])

X = np.around(X, decimals=5)

# Real part of DTFT

plt.title('Real part of DTFT')
plt.xlabel('Omega')
plt.ylabel('Magnitude')
plt.stem(omega, X.real, use_line_collection = True) 
plt.show()

# Imaginary part of DTFT

plt.title('Imaginary part of DTFT')
plt.xlabel('Omega')
plt.ylabel('Magnitude')
plt.stem(omega, X.imag, use_line_collection = True) 
plt.show()

# Magnitude Spectrum

# magnitude = sqrt(A^2 + B^2)
magnitude = np.sqrt((X.real)*(X.real) + (X.imag)*(X.imag))

plt.title('Magnitude Spectrum')
plt.xlabel('Omega')
plt.ylabel('Magnitude')
plt.stem(omega, magnitude, use_line_collection = True) 
plt.show()

# Phase Spectrum

# phase = arctan(B/A)
phase = np.angle(X)

plt.title('Phase Spectrum')
plt.xlabel('Omega')
plt.ylabel('Radians')
plt.stem(omega, phase, use_line_collection = True) 
plt.show()
````

````
import numpy as np
import matplotlib.pyplot as plt
import wave
````

##### a) Plotting the signals in the time domain.

````
alarm = wave.open("alarm.wav", "r")
armageddon = wave.open("armageddon.wav", "r")
cry = wave.open("cry.wav", "r")
wow = wave.open("wow.wav", "r")

# plotting all the audio signals in time domain
plots, audio = plt.subplots(2, 2)

# alarm
alarm_signal = alarm.readframes(-1)
alarm_signal = np.frombuffer(alarm_signal, dtype=np.uint8)
frame_rate = alarm.getframerate()
x = np.linspace(0, len(alarm_signal) / frame_rate, num=len(alarm_signal))

audio[0, 0].set_title('Alarm')
audio[0, 0].set(xlabel='Time', ylabel='Amplitude')
audio[0, 0].plot(x, alarm_signal) 

# armageddon
armageddon_signal = armageddon.readframes(-1)
armageddon_signal = np.frombuffer(armageddon_signal, dtype=np.uint8)
frame_rate = armageddon.getframerate()
x = np.linspace(0, len(armageddon_signal) / frame_rate, num=len(armageddon_signal))

audio[0, 1].set_title('Armageddon')
audio[0, 1].set(xlabel='Time', ylabel='Amplitude')
audio[0, 1].plot(x, armageddon_signal) 

# cry
cry_signal = cry.readframes(-1)
cry_signal = np.frombuffer(cry_signal, dtype=np.uint8)
frame_rate = cry.getframerate()
x = np.linspace(0, len(cry_signal) / frame_rate, num=len(cry_signal))

audio[1, 0].set_title('Cry')
audio[1, 0].set(xlabel='Time', ylabel='Amplitude')
audio[1, 0].plot(x, cry_signal) 

# wow
wow_signal = wow.readframes(-1)
wow_signal = np.frombuffer(wow_signal, dtype=np.uint8)
frame_rate = wow.getframerate()
x = np.linspace(0, len(wow_signal) / frame_rate, num=len(wow_signal))

audio[1, 1].set_title('Wow')
audio[1, 1].set(xlabel='Time', ylabel='Amplitude')
audio[1, 1].plot(x, wow_signal) 

plots.tight_layout()

````

##### b) Computing the DTFT of each of these signals and plotting their magnitude and phase spectra.

````
# alarm 
n = alarm_signal.size

samples = 100
omega = np.linspace(-np.pi, np.pi, samples)
X = np.zeros(samples, dtype=complex)
for w in range(samples):
  wn = omega[w] * (np.arange(n))
  real = (alarm_signal * np.cos(wn)).sum()
  imag = (alarm_signal * np.sin(wn)).sum()

  # X(e^j*omega[w]) = SUM(x * cos(omega[w])) - j *  SUM(x * sin(omega[w])) = A - jB
  X[w] = real - imag * np.array([1j])

magnitude = np.sqrt((X.real)*(X.real) + (X.imag)*(X.imag))
phase = np.angle(X)

plots, alarm = plt.subplots(2, 1)

alarm[0].set_title('Magnitude Spectrum of alarm.wav')
alarm[0].set(xlabel='Omega', ylabel='Magnitude')
alarm[0].stem(omega, magnitude, use_line_collection = True) 

alarm[1].set_title('Phase Spectrum of alarm.wav')
alarm[1].set(xlabel='Omega', ylabel='Radians')
alarm[1].stem(omega, phase, use_line_collection = True) 

plots.tight_layout()
````
